```python
import numpy as np
import pandas as pd
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
from tensorflow import keras
```

Prepare Data

```python
data_raw = pd.read_csv("/content/drive/MyDrive/Colab/Data1/train.csv")
Test_raw = pd.read_csv("/content/drive/MyDrive/Colab/Data1/test.csv")
```

```python
data_raw = data_raw.dropna()
Test_raw = Test_raw.dropna()
```

```python
def Xformat(Data): #reshape and scope data
  return np.array(Data).reshape((len(Data),28,28,1))/255
```

```python
target = Xformat(Test_raw)
```

```python
Yt_raw =  data_raw["label"][:40000]
Xt_raw = data_raw.drop(labels= "label",axis=1)[:40000]
Ytest_raw =  data_raw["label"][40000:]
Xtest_raw = data_raw.drop(labels= "label",axis=1)[40000:]
```

```python
Xt = Xformat(Xt_raw)
Yt = np.array(Yt_raw)
Xtest = Xformat(Xtest_raw)
Ytest = np.array(Ytest_raw)
```

Create model

```python
model = keras.Sequential()
model.add(layers.Conv2D(6,(5,5),input_shape=(28,28,1),activation="relu"))
model.add(layers.MaxPool2D())
model.add(layers.Conv2D(16,(5,5),activation="relu"))
model.add(layers.MaxPool2D())
model.add(layers.Flatten())
model.add(layers.Dense(120,activation="relu"))
model.add(layers.Dense(84,activation="relu"))
model.add(layers.Dense(10,activation="softmax"))
model.compile(loss = "sparse_categorical_crossentropy",optimizer="sgd")
```

```python
keras.utils.plot_model(model,show_shapes=True)
```

```python
model.fit(Xt,Yt,epochs=20)
```

```python
Ztest = model.predict(Xtest)
Ztest = Ztest.argmax(axis=1)
```

```python
np.sum(Ztest != Ytest)/(len(Ytest))
```

```python
results = model.predict(target)
results = results.argmax(axis=1)
```

```python
results = pd.Series(results,name = "Label")
index = pd.Series(range(1,len(results)+1),name = "ImageId")
```
